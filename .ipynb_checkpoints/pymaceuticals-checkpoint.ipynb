```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = pd.read_csv("mouse_drug_data.csv")
clinical_trial_data_to_load = pd.read_csv("clinicaltrial_data.csv")

# Read the Mouse and Drug Data and the Clinical Trial Data
#mouse_drug_data_to_load.head()
#clinical_trial_data_to_load.head()
#clinical_trial_data_to_load.count()
#mouse_drug_data_to_load.count()

# Combine the data into a single dataset
mouse_clinical_trial = pd.merge(clinical_trial_data_to_load,mouse_drug_data_to_load,on="Mouse ID")

# Display the data table for preview
mouse_clinical_trial.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
#mouse_clinical_trial.count()
mouse_clinical_trial_mean_tumor_vol = mouse_clinical_trial[["Drug","Timepoint","Tumor Volume (mm3)"]]
mouse_clinical_trial_mean_tumor_vol["Timepoint"] = mouse_clinical_trial_mean_tumor_vol["Timepoint"].astype(object)

# Convert to DataFrame
mouse_clinical_trial_mean_tumor_vol.groupby(["Drug","Timepoint"]).mean()


# Preview DataFrame
mouse_clinical_trial_mean_tumor_vol.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
mouse_clinical_trial_std_err = mouse_clinical_trial_mean_tumor_vol[["Drug","Timepoint","Tumor Volume (mm3)"]]

# Convert to DataFrame
mouse_clinical_trial_std_err.groupby(["Drug","Timepoint"]).sem().head()
# Preview DataFrame

```

```python
# Minor Data Munging to Re-Format the Data Frames
by_timepoint = mouse_clinical_trial_mean_tumor_vol.set_index(["Timepoint"])
by_timepoint = by_timepoint.pivot_table("Tumor Volume (mm3)",["Timepoint"],"Drug")
by_timepoint.head()
# Preview that Reformatting worked
```

```python
# Generate the Plot (with Error Bars)
x_axis = by_timepoint.index.tolist()
#x_axis
capomulin = by_timepoint["Capomulin"]
infubinol = by_timepoint["Infubinol"]
ketapril = by_timepoint["Ketapril"]
placebo = by_timepoint["Placebo"]
plt.plot(x_axis,capomulin,marker ='o', color='red', label="Capomulin", linewidth=0.5)
plt.plot(x_axis,infubinol,marker ='^', color='blue', label="Infubinol", linewidth=0.5)
plt.plot(x_axis,ketapril,marker ='s', color='green', label="Ketapril",linewidth=0.5)
plt.plot(x_axis,placebo,marker ='D', color='black', label="Placebo",linewidth=0.5)
plt.legend(loc="upper left")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.grid()
# Save the Figure
plt.savefig("Tumor_Response_to_Treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_response = mouse_clinical_trial[["Drug","Timepoint","Metastatic Sites"]]
metastatic_response["Timepoint"] = metastatic_response["Timepoint"].astype(object)
# Convert to DataFrame
metastatic_response.groupby(["Drug","Timepoint"]).mean()
# Preview DataFrame
metastatic_response.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_response_std_err = metastatic_response[["Drug","Timepoint","Metastatic Sites"]]


# Convert to DataFrame
metastatic_response_std_err.groupby(["Drug","Timepoint"]).sem().head()
# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
by_msite = metastatic_response.set_index(["Timepoint"])
by_msite = by_msite.pivot_table("Metastatic Sites",["Timepoint"],"Drug")
by_msite.head()
# Preview that Reformatting worked
```

```python
# Generate the Plot (with Error Bars)
x_axis2 = by_msite.index.tolist()
#x_axis
capomulin2 = by_msite["Capomulin"]
infubinol2 = by_msite["Infubinol"]
ketapril2 = by_msite["Ketapril"]
placebo2 = by_msite["Placebo"]
plt.plot(x_axis2,capomulin2,marker ='o', color='red', label="Capomulin", linewidth=0.5)
plt.plot(x_axis2,infubinol2,marker ='^', color='blue', label="Infubinol", linewidth=0.5)
plt.plot(x_axis2,ketapril2,marker ='s', color='green', label="Ketapril",linewidth=0.5)
plt.plot(x_axis2,placebo2,marker ='D', color='black', label="Placebo",linewidth=0.5)
plt.legend(loc="upper left")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.title("Metastatic Spread During Treatment")
plt.grid()
# Save the Figure
plt.savefig("Metastatic_Response_to_Treatment.png")
# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survivors = mouse_clinical_trial.groupby(["Drug","Timepoint"]).count()
# Convert to DataFrame
del survivors["Tumor Volume (mm3)"]
del survivors["Metastatic Sites"]
# Preview DataFrame
survivors.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survivors = survivors.pivot_table("Mouse ID",["Timepoint"],"Drug")
survivors.head()
# Preview the Data Frame
```

```python
# Generate the Plot (Accounting for percentages)
x_axis3 = survivors.index.tolist()
#x_axis
capomulin3 = survivors["Capomulin"]
infubinol3 = survivors["Infubinol"]
ketapril3 = survivors["Ketapril"]
placebo3 = survivors["Placebo"]
plt.plot(x_axis3,capomulin3,marker ='o', color='red', label="Capomulin", linewidth=0.5)
plt.plot(x_axis3,infubinol3,marker ='^', color='blue', label="Infubinol", linewidth=0.5)
plt.plot(x_axis3,ketapril3,marker ='s', color='green', label="Ketapril",linewidth=0.5)
plt.plot(x_axis3,placebo3,marker ='D', color='black', label="Placebo",linewidth=0.5)
plt.legend(loc="lower left")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.title("Metastatic Spread During Treatment")
plt.grid()
# Save the Figure
plt.savefig("Survival_Rates.png")
# Show the Figure
plt.show()
```
